```python
 import pandas as pd 
 import numpy as np
```

```python
data= pd.read_csv("/content/drive/MyDrive/data.csv")
```

```python
data
```

```python
data.drop(labels=data.columns[0],axis=1,inplace=True)
```

```python
data.drop(labels=data.columns[-1],axis=1,inplace=True)
```

```python
neg_class_data = data[data[data.columns[0]]=='B']
```

```python
neg_class_data
```

```python
pos_class_data = data[data[data.columns[0]]=='M']
```

```python
pos_class_data
```

```python
trainingData= 0.7
cvData= 0.2
testingData= 0.1
```

```python
trainingDatalen= int(trainingData*data.shape[0])
cvdatalen=int(cvData*data.shape[0])
testingDatalen=int(testingData*data.shape[0])
```

```python
trainingDatalen
```

```python
#balancing dataset
neg_class_trainingdata=neg_class_data.iloc[0:trainingDatalen//2,:]
pos_class_trainingdata=pos_class_data.iloc[0:trainingDatalen//2,:]
```

```python
neg_class_trainingdata
```

```python
pos_class_trainingdata
```

```python
TRAININGDATA=pd.concat([neg_class_trainingdata,pos_class_trainingdata], axis=0)
#axis is by deafault 0 so that the data is concatenated row wise instead of column wise
```

```python
TRAININGDATA
```

```python
remaining_neg=neg_class_data.iloc[trainingDatalen//2:,:]
remaining_pos=pos_class_data.iloc[trainingDatalen//2:,:]
```

```python
remaining_neg
```

```python
remaining_pos
```

```python
RemainingData=pd.concat([remaining_neg,remaining_pos])
```

```python
RemainingData
```

```python
cvdatalen
```

```python
#shuffling of data(row wise) as both types of tumor are present in all 3 types of data
#starting with extracting their index and then shuffling it 
remainingdataindex=np.arange(0,RemainingData.shape[0])
```

```python
np.random.shuffle(remainingdataindex)
```

```python
remainingdataindex
```

```python
RemainingData=RemainingData.iloc[remainingdataindex]
```

```python
RemainingData
```

```python
cvData = RemainingData.iloc[0:cvdatalen,:]
testingData = RemainingData.iloc[cvdatalen:,]
cvData[cvData[cvData.columns[0]]=='M']
```

```python
testingData[testingData[testingData.columns[0]]=='M']
#cv data and testing data have some amount of M data 
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
obj = GaussianNB()
```

```python
obj.fit(X=TRAININGDATA.iloc[:,1:],y=TRAININGDATA.iloc[:,0])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_cv_pred = obj.predict(cvData.iloc[:,1:])
```

```python
y_cv_pred
```

```python
confusion_matrix(y_true=cvData.iloc[:,0],y_pred=y_cv_pred)
```

```python
######################## LOGISTIC REGRESSION #######################



from sklearn.metrics import classification_report
```

```python
print(classification_report(y_true=cvData.iloc[:,0],y_pred=y_cv_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression_algo = LogisticRegression(max_iter=10000)
```

```python
logistic_regression_algo.fit(TRAININGDATA.iloc[:,1:],y=TRAININGDATA.iloc[:,0])
```

```python
y_cv_pred = logistic_regression_algo.predict(cvData.iloc[:,1:])
```

```python
y_cv_pred
```

```python
confusion_matrix(y_true=cvData.iloc[:,0],y_pred=y_cv_pred)
```

```python
print(classification_report(y_true=cvData.iloc[:,0],y_pred=y_cv_pred))
```
