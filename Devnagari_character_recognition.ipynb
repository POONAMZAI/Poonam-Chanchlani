```python
#PARELLEL PROCESSING LIBRARY : FULL utilization of the threads as in decreasing the time takes to do one task
from concurrent.futures import ThreadPoolExecutor
```

```python
import numpy as np
```

```python
#pyspark, keras and pytorch are a must learn 
```

```python
# FUNCTION 1

# we made a function image_flattening
# it will convert the square matrix into a rectangle one 
def image_flattening(image_np_array):
#earlier the shape was (32,32), this function will convert it into (1024,)
  return image_np_array.reshape(image_np_array.shape[0]*image_np_array.shape[1],)
```

```python
# FUNCTION 2

# this function will read the image
def read_image(img_path):
  return plt.imread(img_path)
```

```python
#FUNCTION 3

# ths unction will make a list of 1700 paths of the images in every single folder 
def single_dir_image_path_list(base_train_dir_path):
#this image_list contains the listof1700 images
    image_list = os.listdir(base_train_dir_path)
# 4 workers will work on these 46 folders unlike forloop which takes more time
# lambda functions are one liner functions
    with ThreadPoolExecutor(max_workers=4) as pool:
            image_path_list = list(pool.map(lambda x: base_train_dir_path+"/"+x,image_list))
    return image_path_list
#output will be a list : starting with a string(the path of the first image)...1700 paths will be stacked on one another
```

```python
# FUNCTION 4

def single_category_dir_path(base_train_path):
  #list of 46 folders will be stored in dirs
  dirs=os.listdir(base_train_path)
  dir_path_list = list()
  for dir_name in dirs:
    dir_path_list.append(base_train_path+"/"+dir_name)
  return dir_path_list
  #this will return the 46 paths of the folders stacked
```

```python
 dir_path_list = single_category_dir_path(base_train_path)
```

```python
dir_path_list
#46 paths list
```

```python
image_path_list = single_dir_image_path_list(dir_path_list[0])
image_path_list
```

```python
# FUNCTION 5

# list of arrays of1700 images
def single_dir_images_np_array_list(image_path_list):
  with ThreadPoolExecutor(max_workers=4) as pool:
    images_np_array_list = list(pool.map(read_image,image_path_list))
  return images_np_array_list
```

```python
images_np_array_list = single_dir_images_np_array_list(image_path_list)
```

```python
from dummy_threading import Thread
# FUNCTION 6


def single_dir_images_matrix(images_np_array_list):
  with ThreadPoolExecutor(max_workers=4) as pool:
    reshaped_images_list = list(pool.map(image_flattening,images_np_array_list))
  return np.array(reshaped_images_list)
```

```python
reshaped_images_matrix = single_dir_images_matrix(images_np_array_list)
                                          
```

```python
reshaped_images_matrix.shape
```

```python
#this will reshape the matrices and every functions' output is an input for the next function(pipeline)
reshaped_images_matrix_list = list()
for single_dir_path in dir_path_list:
  image_path_list = single_dir_image_path_list(single_dir_path)
  images_np_array_list = single_dir_images_np_array_list(image_path_list)
  reshaped_images_matrix = single_dir_images_matrix(images_np_array_list)
  reshaped_images_matrix_list.append(reshaped_images_matrix)
```

```python
#to import the outputs in the 
import pandas as pd
```

```python
training_data_input_features = np.concatenate(reshaped_images_matrix_list)
```

```python
training_data = pd.DataFrame(data=training_data_input_features,columns=["Pixel_"+str(i) for i in range(1,training_data_input_features.shape[1]+1)])
```
